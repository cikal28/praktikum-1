```python
#bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
data = [5, 2, 8, 1]
print("Sebelum sorting:", data)
hasil = bubble_sort(data)
print("Sesudah sorting:", hasil)
```

```python
Data sebelum sorting: [5, 2, 8, 1]

Iterasi ke- 1
Membandingkan 5 dan 2
Ditukar menjadi: [2, 5, 8, 1]

Membandingkan 5 dan 8
Tidak ditukar

Membandingkan 8 dan 1
Ditukar menjadi: [2, 5, 1, 8]

Iterasi ke- 2
Membandingkan 2 dan 5
Tidak ditukar

Membandingkan 5 dan 1
Ditukar menjadi: [2, 1, 5, 8]

Iterasi ke- 3
Membandingkan 2 dan 1
Ditukar menjadi: [1, 2, 5, 8]

Iterasi ke- 4
Data setelah sorting: [1, 2, 5, 8]

Bubble Sort melakukan:
6 kali perbandingan
4 kali pertukaran

kompleksitas waktunya O(n²) karena menggunakan 2 perulangan untuk
 membandingkan setia elemen
```

```python
# Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
data = [5, 2, 8, 1]
print("Sebelum sorting:", data)
hasil = selection_sort(data)
print("Sesudah sorting:", hasil)
```

```python
#insertion sort
# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
data = [5, 2, 8, 1]
print("Sebelum sorting:", data)
hasil = insertion_sort(data)
print("Sesudah sorting:", hasil)
```

```python
Data sebelum sorting: [5, 2, 8, 1]

Iterasi ke- 1
Membandingkan 5 dan 2
Data digeser: [5, 5, 8, 1]
Hasil sementara: [2, 5, 8, 1]

Iterasi ke- 2
Hasil sementara: [2, 5, 8, 1]

Iterasi ke- 3
Membandingkan 8 dan 1
Data digeser: [2, 5, 8, 8]

Membandingkan 5 dan 1
Data digeser: [2, 5, 5, 8]

Membandingkan 2 dan 1
Data digeser: [2, 2, 5, 8]

Hasil sementara: [1, 2, 5, 8]

Data setelah sorting: [1, 2, 5, 8]

Total perbandingan: 4
Total pergeseran: 4
Jumlah data = 4
Kompleksitas Insertion Sort = O(n²)
```

```python
#merge sort
# Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]
        merge_sort(left)
        merge_sort(right)
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
    return arr
data = [5, 2, 8, 1]
print("Sebelum sorting:", data)
hasil = merge_sort(data)
print("Sesudah sorting:", hasil)
```
