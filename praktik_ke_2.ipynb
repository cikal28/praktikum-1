```python
def find_maximum (T):
  m = len(T)
  for i in range(m - 1, -1 ,-1):
    if T[i] is not None:
      return i
  return None

def find_maximum_value(T):
  m = len(T)
  for i in range(m - 1 , -1 ,-1):
    if T[i] is not None:
      return T[i]
  return None

T = [None, 5 , None , 9 , None, None]

print(find_maximum(T))
print(find_maximum_value(T))
```
