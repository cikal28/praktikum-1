```python
#fungsi Algoritma PCA
data = [
    [2, 3],
    [3, 5],
    [4, 6],
    [5, 8]]
mean = []
for kolom in range(len(data[0])):
    jumlah = 0
    for baris in data:
        jumlah += baris[kolom]
    mean.append(jumlah / len(data))
print("Mean:", mean)

center = []
for baris in data:
    temp = []
    for i in range(len(baris)):
        temp.append(baris[i] - mean[i])
    center.append(temp)
print("Data Center:")
for i in center:
    print(i)

transpose = []
for i in range(len(center[0])):
    temp = []
    for j in range(len(center)):
        temp.append(center[j][i])
    transpose.append(temp)

cov = []
for i in range(len(transpose)):
    row = []
    for j in range(len(center[0])):
        total = 0
        for k in range(len(center)):
            total += transpose[i][k] * center[k][j]
        row.append(total / (len(center) - 1))
    cov.append(row)
print("Covariance Matrix:")
for i in cov:
    print(i)
```
