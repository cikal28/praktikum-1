```python
def total(angka ,n):
  jumlah = 0
  for i in range (n):
    jumlah = jumlah + angka [i]
  return jumlah
angka = [5,3,7,9]
n = len (angka)
print(total (angka,n))
```

```python
a= 3
b= 2
angka = [(3,2,7),(2,6,8),(5,1,9)]
total = 0

for i in range(a):
  for j in range(b):
    total += angka [i][j]
print (total)
```
