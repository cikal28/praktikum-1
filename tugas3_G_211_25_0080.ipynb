coding


```python
from itertools import permutations

jarak = {
    ('a','b'):4,
    ('a','c'):2,
    ('b','c'):10
}

kota = ['b','c']

minimum = float('inf')
terbaik = ()

for rute in permutations(kota):
    total = 0
    awal = 'a'

    for tujuan in rute:
        if (awal, tujuan) in jarak:
            total += jarak[(awal, tujuan)]
        else:
            total += jarak[(tujuan, awal)]
        awal = tujuan

    if total < minimum:
        minimum = total
        terbaik = rute

print("Rute Terpendek : a ->", " -> ".join(terbaik))
print("Total Jarak    :", minimum)
```

Analisa kompleksitas waktu

```python
 Algoritma menggunakan metode faktorial dengan mencoba seluruh kemungkinan rute sebanyak n!.
Setiap rute dihitung total jaraknya sebanyak n langkah.
Maka kompleksitas waktu adalah O(n × n!).

t(n)= n x n!
O(n × n!)
```
